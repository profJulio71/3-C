### Imports de dependências

```python
import numpy as np
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense, Input, Dropout
from tensorflow.keras.metrics import Precision, Recall
from tensorflow.keras.models import Model
import tensorflow as tf
import matplotlib.pyplot as plt
```

### Configurações

```python
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
```

### Leitura dos dados

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
# Reshape dos dados para incluir a dimensão do canal (1 para imagens em escala de cinza)
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32')
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32')

# Normalização dos valores dos pixels para o intervalo 0-1
X_train = X_train / 255.0
X_test = X_test / 255.0
```

### Ajuste de classes para one-hot encoding

```python
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)
```

### Criação o modelo

```python
model = Sequential()

# Primeira camada convolucional
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(BatchNormalization())  # Normalização em lote para melhorar a estabilidade
model.add(MaxPooling2D((2, 2)))  # Camada de pooling para redução de dimensionalidade

# Segunda camada convolucional
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(BatchNormalization())  # Normalização em lote
model.add(MaxPooling2D((2, 2)))  # Outra camada de pooling

# Flatten (achatamento) das saídas das camadas convolucionais para usar em uma camada densa
model.add(Flatten())

# Camada totalmente conectada com função de ativação relu
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

# Camada totalmente conectada com função de ativação relu
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))

# Camada de saída (10 classes para os 10 itens de roupas)
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy',    metrics=['accuracy', Precision(), Recall()])
```

### Treinamento do modelo

```python
history = model.fit(X_train, y_train, epochs=6, batch_size=128, validation_data=(X_test, y_test), )
```

### Apresentração de uma Imagem de exemplo e exemplos de saída de algumas camadas do modelo

```python
# Função para visualizar as saídas das camadas intermediárias
def visualize_layer_outputs(model, image):
    
    layer_outputs = [layer.output for layer in model.layers if isinstance(layer, (tf.keras.layers.Conv2D, tf.keras.layers.MaxPooling2D))]
    intermediate_model = Model(inputs=model.input, outputs=layer_outputs)

    intermediate_outputs = intermediate_model.predict(image[np.newaxis, ...])

    for i, (feature_maps, layer) in enumerate(zip(intermediate_outputs, layer_outputs)):
        layer_type = 'Convolucional' if isinstance(model.layers[i], tf.keras.layers.Conv2D) else 'Pooling'
        print(f"Visualizando a saída da camada {i+1} ({layer_type}) com forma {feature_maps.shape}")
        
        num_feature_maps = feature_maps.shape[-1]
        
        plt.figure(figsize=(12, 12))
        for j in range(min(num_feature_maps, 4)):
            plt.subplot(4, 4, j + 1)
            plt.imshow(feature_maps[0, :, :, j], cmap='gray')
            plt.axis('off')
        plt.show()

# Escolher uma imagem do conjunto de treinamento
image = X_train[1]

# Exibir a imagem usando matplotlib
plt.imshow(image, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')
plt.show()

# Visualizar as saídas das camadas convolucionais e pooling
visualize_layer_outputs(model, image)
```
