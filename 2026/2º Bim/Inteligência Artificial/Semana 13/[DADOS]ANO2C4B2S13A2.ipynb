### Imports de dependГЄncias

```python
import pandas as pd
import spacy
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import RSLPStemmer
import re
```

### ConfiguraГ§Гµes

```python
pd.set_option("display.max_colwidth", None)  # Exibe o conteГєdo completo das cГ©lulas
pd.set_option("display.max_rows", None)  # Exibe todas as linhas sem truncar
```

### Leitura dos dados

```python
dados = {
    "descricao": [
        "Compra no supermercado Eee",
        "Pagamento de conta de luz",
        "TransferГЄncia recebida",
        "Assinatura mensal streaming Nnn",
        "Jantar no restaurante italiano",
        "Compra de gasolina no posto Sss",
        "Pagamento de aluguel",
        "Compra de roupas na loja Zzz",
        "Compra de eletrГґnicos no Mmm",
        "Compra de livros na Aaa",
        "Compra de remГ©dios na farmГЎcia",
        "Pagamento de fatura do cartГЈo",
        "Compra de mГіveis para a casa",
        "Assinatura de mГєsica Sss",
        "TransferГЄncia enviada para JoГЈo"
    ]
}

df = pd.DataFrame(dados)
```

### TransformaГ§Гµes bГЎsicas do texto

```python
def preprocessamento(texto):
    texto = re.sub(r'[^\w\s]', '', texto)
    return texto.lower()

df['descricao'] = df['descricao'].apply(preprocessamento)
```

### ApresentaГ§ГЈo dos dados

```python
df
```

### TokenizaГ§ГЈo

```python
df['tokens'] = df['descricao'].apply(word_tokenize)
df[['descricao', 'tokens']]
```

### RemoГ§ГЈo de Stopwords

```python
stop_words_pt = set(stopwords.words('portuguese'))
df['tokens_sem_stopwords'] = df['tokens'].apply(lambda tokens: [word for word in tokens if word not in stop_words_pt])
df[['tokens', 'tokens_sem_stopwords']]
```

### Bag of Words

```python
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(df['descricao'])
pd.DataFrame(bag_of_words.toarray(), columns=vectorizer.get_feature_names_out())
```

### Stemming

```python
# Stemming com RSLPStemmer em portuguГЄs
stemmer = RSLPStemmer()
df['tokens_stemmed'] = df['tokens_sem_stopwords'].apply(lambda tokens: [stemmer.stem(word) for word in tokens])
df[['tokens_sem_stopwords', 'tokens_stemmed']]
```

### LematizaГ§ГЈo

```python
nlp = spacy.load('pt_core_news_sm')
# LematizaГ§ГЈo com Spacy para portuguГЄs
def lematizacao(tokens):
    doc = nlp(" ".join(tokens))
    return [token.lemma_ for token in doc]

df['tokens_lemmatized'] = df['tokens_sem_stopwords'].apply(lematizacao)
df[['tokens_sem_stopwords', 'tokens_lemmatized']]
```
