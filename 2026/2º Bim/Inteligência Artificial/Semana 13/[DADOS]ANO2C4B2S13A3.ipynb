```python
import re
import pandas as pd
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import nltk
```

### Configurações

```python
pd.set_option("display.max_colwidth", None)  # Exibe o conteúdo completo das células
pd.set_option("display.max_rows", None)  # Exibe todas as linhas sem truncar
```

### Leitura dos dados

```python
dados = {
    "descricao_original": [
        "Compra no supermercado Eee",
        "Pagamento de conta de luz",
        "Transferência recebida",
        "Assinatura mensal streaming Nnn",
        "Jantar no restaurante italiano",
        "Compra de gasolina no posto Sss",
        "Pagamento de aluguel",
        "Compra de roupas na loja Zzz",
        "Compra de eletrônicos no Mmm",
        "Compra de livros na Aaa",
        "Compra de remédios na farmácia",
        "Pagamento de fatura do cartão",
        "Compra de móveis para a casa",
        "Assinatura de música Sss",
        "Transferência enviada para João"
    ]
}

df = pd.DataFrame(dados)
```

### Transformações básicas do texto

```python
def preprocessamento(texto):
    texto = re.sub(r'[^\w\s]', '', texto)
    return texto.lower()

df['descricao'] = df['descricao_original'].apply(preprocessamento)
```

### Apresentação dos dados

```python
df
```

### Tokenização

```python
df['tokens'] = df['descricao'].apply(word_tokenize)
df[['descricao', 'tokens']]
```

### Criação e treinamento do modelo Word2Vec

```python
#TO DO
# Configurações: vetor de 10 dimensões, janela de contexto de 2 palavras, e 5 iterações de treinamento
model = Word2Vec(sentences=df['tokens'], vector_size=10, window=2, min_count=1, epochs=5)
#model = Word2Vec(sentences=df['tokens'], vector_size=30, window=2, min_count=1, epochs=5)
```

### Função para gerar vetores médios das palavras em uma descrição

```python
def obter_embedding_medio(tokens, model):
    # Gera a média dos embeddings de cada token presente no modelo
    vetor_palavras = [model.wv[word] for word in tokens if word in model.wv]
    if len(vetor_palavras) > 0:
        return sum(vetor_palavras) / len(vetor_palavras)
    else:
        return [0] * model.vector_size
```

### Geração embeddings para cada transação

```python
df['embedding'] = df['tokens'].apply(lambda tokens: obter_embedding_medio(tokens, model))
```

```python
df
```

### Calculo de similaridade entre transações

```python
similaridade = cosine_similarity(list(df['embedding']))
```

### Apresentação dos resultados

```python
# Função para exibir as três transações mais semelhantes para uma transação específica
def exibir_similares(transacao, df, similaridade):
    idx = df.index[df['descricao_original'] == transacao].tolist()[0]
    similaridades = list(enumerate(similaridade[idx]))
    similares_ordenados = sorted(similaridades, key=lambda x: x[1], reverse=True)
    print(f"\nTransação: {transacao}")
    for i, score in similares_ordenados[1:4]:  # Mostra as 3 mais semelhantes
        print(f"- Similar a: '{df.iloc[i]['descricao']}' (similaridade: {score:.2f})")
```

```python
# Exemplos de similaridade
exibir_similares('Assinatura mensal streaming Nnn', df, similaridade)
exibir_similares('Transferência recebida', df, similaridade)
```
